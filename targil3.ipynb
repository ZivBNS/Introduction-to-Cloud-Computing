```python
from firebase import firebase
import time

# Initialize Firebase connection using your database URL
FBconn = firebase.FirebaseApplication('https://test-538ee-default-rtdb.europe-west1.firebasedatabase.app/', None)

def get_all_records():
    """Retrieve all word records from the database"""
    return FBconn.get('/wordAnalysis/', None)

def update_word_count(record_id, new_count):
    """Update the count of an existing word record"""
    return FBconn.put('/wordAnalysis/' + record_id, 'Count', new_count)

def delete_word_record(record_id):
    """Delete a word record by its ID"""
    return FBconn.delete('/wordAnalysis/', record_id)

def find_record_id_by_word(word_to_find):
    """Helper to find the record ID for a specific word"""
    records = get_all_records()
    if records:
        for record_id, data in records.items():
            if data['Word'] == word_to_find.lower():
                return record_id, data['Count']
    return None, 0

def display_all_words():
    """Format and display all saved words and their counts"""
    records = get_all_records()
    if records:
        print("\nAnalysis Results:")
        results = [f"{data['Word']} :{data['Count']}" for record_id, data in records.items()]
        print(", ".join(results))
    else:
        print("\nNo words found in database.")

while True:
    print("\n--- Word Counter Firebase Menu ---")
    print("1. Add a single word")
    print("2. Add full text for analysis")
    print("3. Update word count manually")
    print("4. Delete a word")
    print("5. View all saved words")
    print("6. Exit")

    choice = input("\nEnter your choice (1-6): ")

    if choice == '1':
        # Add single word - check if exists first
        word = input("Enter word: ").strip().lower()
        if word:
            record_id, current_count = find_record_id_by_word(word)
            if record_id:
                update_word_count(record_id, current_count + 1)
                print(f"Count updated for '{word}'.")
            else:
                data = {'Word': word, 'Count': 1}
                FBconn.post('/wordAnalysis/', data)
                print(f"New word '{word}' added.")

    elif choice == '2':
        # Add full text - parse and update database
        text = input("Enter full text: ").strip().lower()
        words_list = text.split()
        for w in words_list:
            record_id, current_count = find_record_id_by_word(w)
            if record_id:
                update_word_count(record_id, current_count + 1)
            else:
                data = {'Word': w, 'Count': 1}
                FBconn.post('/wordAnalysis/', data)
        print("Text analyzed and database updated.")

    elif choice == '3':
        # Manual update
        display_all_words()
        word_to_upd = input("\nEnter the word you want to update: ").strip().lower()
        record_id, _ = find_record_id_by_word(word_to_upd)
        if record_id:
            try:
                new_count = int(input(f"Enter new count for '{word_to_upd}': "))
                update_word_count(record_id, new_count)
                print("Count updated successfully.")
            except ValueError:
                print("Invalid input. Please enter a number.")
        else:
            print("Word not found.")

    elif choice == '4':
        # Delete record
        display_all_words()
        word_to_del = input("\nEnter the word to delete: ").strip().lower()
        record_id, _ = find_record_id_by_word(word_to_del)
        if record_id:
            delete_word_record(record_id)
            print(f"'{word_to_del}' deleted.")
        else:
            print("Word not found.")

    elif choice == '5':
        display_all_words()

    elif choice == '6':
        print("Closing program...")
        break

    else:
        print("Invalid choice, please try again.")
```
